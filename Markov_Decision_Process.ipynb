```python
import copy
import timeit
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
from ipywidgets import interactive
import ipywidgets as widgets
```

Create the following grid world.

**States:** A 10x10 grid

**Actions:** Up, Down, Left, Right

**Tranistion probabilities:**
* 0.7 in the direction of action
* 0.1 in the three other directions
* The robot bounces back to the same state near edges

**Rewards:**
* (7,8) has a reward +10
* (2,7) has a reward +3
* (4,3) has a reward -5
* (7,3) has a reward -10
* No reward in other states

This example is based on Decision Making Under Uncertainty by M.J. Kochenderfer.

```python
#Let's define MDP paras
def createGrid10World():
    def xy2s(y, x):
        x = max(x, 0)
        y = max(y, 0)
        x = min(x, 9)
        y = min(y, 9)
        out = np.ravel_multi_index(np.array([x,y]), (10,10))
        return out

    def s2xy(s):
        x, y = np.unravel_index(s, (10,10))
        return y, x

    def gridPlot(ax, im, title='', cmap='Blues'):
        pl.imshow(im, interpolation='none', cmap=cmap, origin='lower')
        pl.colorbar()
        ax.set_xticks(np.arange(0, 10, 1));
        ax.set_yticks(np.arange(0, 10, 1));
        ax.set_xticklabels(np.arange(0, 10, 1));
        ax.set_yticklabels(np.arange(0, 10, 1));
        ax.set_xticks(np.arange(-.5, 10, 1), minor=True);
        ax.set_yticks(np.arange(-.5, 10, 1), minor=True);
        ax.grid(which='minor', color='w', linestyle='-', linewidth=1)
        pl.title(title);
        return

    A = ['left', 'right', 'up', 'down']
    S = np.arange(100)
    T = np.zeros((len(S), len(A), len(S)))
    R = np.zeros((len(S), len(A)))
    for s in S:
        x, y = s2xy(s)
        if x == 2 and y == 7:
            R[s, :] = 3
        elif x == 7 and y == 8:
            R[s, :] = 10
        else:
            if x == 7 and y == 3:
                R[s, :] = -10
            elif x == 4 and y == 3:
                R[s, :] = -5
            elif x == 0:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 0] = -0.7
            elif x == 9:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 1] = -0.7
            elif y == 0:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 3] = -0.7
            elif y == 9:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 2] = -0.7

            for a in A:
                if a == 'left':
                    T[s, 0, xy2s(x - 1, y)] += 0.7
                    T[s, 0, xy2s(x + 1, y)] += 0.1
                    T[s, 0, xy2s(x, y - 1)] += 0.1
                    T[s, 0, xy2s(x, y + 1)] += 0.1
                elif a == 'right':
                    T[s, 1, xy2s(x + 1, y)] += 0.7
                    T[s, 1, xy2s(x - 1, y)] += 0.1
                    T[s, 1, xy2s(x, y - 1)] += 0.1
                    T[s, 1, xy2s(x, y + 1)] += 0.1
                elif a == 'up':
                    T[s, 2, xy2s(x, y + 1)] += 0.7
                    T[s, 2, xy2s(x, y - 1)] += 0.1
                    T[s, 2, xy2s(x - 1, y)] += 0.1
                    T[s, 2, xy2s(x + 1, y)] += 0.1
                elif a == 'down':
                    T[s, 3, xy2s(x, y - 1)] += 0.7
                    T[s, 3, xy2s(x, y + 1)] += 0.1
                    T[s, 3, xy2s(x - 1, y)] += 0.1
                    T[s, 3, xy2s(x + 1, y)] += 0.1

    for a, c_x, c_y in [(0,0,0), (0,0,9),(1,9,0),(1,9,9),(2,0,9),(2,9,9),(3,0,0),(3,9,0)]:
        R[xy2s(c_x,c_y),a] = -0.8

    discount = 0.9

    nextStates = {}
    for si in range(len(S)):
        for ai in range(len(A)):
            nextStates[(si,ai)] = np.where((T[si, ai, :] != 0) == True)[0]

    return {'S':S, 'A':A, 'T':T, 'R':R, 'discount':discount, 'nextStates':nextStates, 'gridPlot':gridPlot, 'xy2s':xy2s, 's2xy':s2xy}
```

```python
class MDP():
    def __init__(self):
        pass

    def inbuilt_init(self, mdp_dict):
        self.S = mdp_dict['S']
        self.A = mdp_dict['A']
        self.T = mdp_dict['T']
        self.R = mdp_dict['R']
        self.discount = mdp_dict['discount']
        self.nextStates = mdp_dict['nextStates']
        self.gridPlot = mdp_dict['gridPlot']
        self.xy2s = mdp_dict['xy2s']
        self.s2xy = mdp_dict['s2xy']
```

```python
#Define the MDP
mdp = MDP()
mdp.inbuilt_init(mdp_dict=createGrid10World())

#Plot states
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=mdp.S.reshape((10,10)), title='States', cmap='Greys')

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.R[:,a].reshape((10,10)), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards - clipped larger +ve values for visualizing the edges', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=np.clip(mdp.R[:,a].reshape((10,10)), -1, -0.5), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
s0_x, s0_y = 3, 5
s0 = mdp.xy2s(s0_x, s0_y)
pl.figure(figsize=(15,3))
pl.suptitle('Transition probabilities T(s1|s0=({},{}),a)'.format(s0_x, s0_y), fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.T[s0,a,:].reshape((10,10)), title='a='+mdp.A[a], cmap='Blues')
pl.show()
```

### 1. Policy evaluation

Computing the utility, U.

$U^\pi_k(s) = R(s, \pi(s)) + \gamma \sum_{s'} T(s' \mid s, \pi(s))U^\pi_{k-1}(s')$

```python
def iterativePolicyEvaluation(mdp, policy, numIterations=10, gamma=0.9):
    U = np.zeros(len(mdp.S))      # initialize utilities
    for t in range(numIterations):
        U_old = copy.copy(U)
        for s in mdp.S:
            a = policy if isinstance(policy, int) else policy[s]
            # transition probabilities for all next states
            probs = mdp.T[s, a]              # shape: (num_states,)
            rewards = mdp.R[s, a]            # shape: (num_states,) or scalar
            U[s] = np.sum(probs * (rewards + gamma * U_old))
    return U

numIterations = 5
pl.figure(figsize=(15,3))
pl.suptitle('Utilities', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    U = iterativePolicyEvaluation(mdp=mdp, policy=a, numIterations=numIterations)
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='a='+mdp.A[a], cmap='jet')
pl.show()
#print(np.round(U.reshape(10,10),3))
```

```python
def f(action, numIter=1):
    U = iterativePolicyEvaluation(mdp, policy=mdp.A.index(action), numIterations=numIter)
    pl.figure(figsize=(3,3))
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='Utility', cmap='jet')
    pl.show()

interactive_plot = interactive(f, action=widgets.ToggleButtons(options=['left', 'right', 'up', 'down']),
                               numIter=widgets.IntSlider(min=0, max=20, step=1, value=0))
interactive_plot
```

```python
#Value iteration
def valueIteration(mdp, numIterations=1, gamma=0.9):
    U = np.zeros(len(mdp.S))
    for t in range(numIterations):
        U_old = copy.copy(U)
        for s in mdp.S:
            action_values = []
            for a in range(len(mdp.A)):   # loop over action indices
                probs = mdp.T[s, a]       # (num_states,)
                rewards = mdp.R[s, a]     # scalar or (num_states,)
                if np.ndim(rewards) == 0: # scalar reward
                    q = rewards + gamma * np.sum(probs * U_old)
                else:                     # reward per next state
                    q = np.sum(probs * (rewards + gamma * U_old))
                action_values.append(q)
            U[s] = max(action_values)
    return U

def policyExtration(mdp, U, gamma=0.9):
    policy = np.zeros(len(mdp.S), dtype=int)
    for s in mdp.S:
        best_a = 0
        best_val = -np.inf
        for a in range(len(mdp.A)):       # loop over action indices
            probs = mdp.T[s, a]
            rewards = mdp.R[s, a]
            if np.ndim(rewards) == 0:
                q = rewards + gamma * np.sum(probs * U)
            else:
                q = np.sum(probs * (rewards + gamma * U))
            if q > best_val:
                best_val = q
                best_a = a
        policy[s] = best_a
    return policy

U = valueIteration(mdp, numIterations=2)
policy = policyExtration(mdp, U=U)
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='Utility', cmap='jet')
for s in range(100):
    x, y = mdp.s2xy(s)
    if policy[s] == 0:
        m='\u02C2'
    elif policy[s] == 1:
        m='\u02C3'
    elif policy[s] == 2:
        m='\u02C4'
    elif policy[s] == 3:
        m='\u02C5'
    pl.text(x-0.5,y-1,m,color='k',size=20)
pl.show()
```

### 2. Policy iteration

Policy evaluation can be used in policy iteration:
1. Given the current policy, compute U
2. Using U, compute a new policy

```python
def policyIteration(mdp, numIterations=1, gamma=0.9, evalIterations=10):
    U_pi_k = np.zeros(len(mdp.S))  # initial utilities
    pi_k = np.random.randint(low=0, high=4, size=len(mdp.S), dtype=int)  # initial random policy
    pi_kp1 = copy.copy(pi_k)

    for t in range(numIterations):
        # --- Policy Evaluation ---
        for _ in range(evalIterations):
            U_old = copy.copy(U_pi_k)
            for s in mdp.S:
                a = pi_k[s]
                probs = mdp.T[s, a]
                rewards = mdp.R[s, a]
                if np.ndim(rewards) == 0:
                    U_pi_k[s] = rewards + gamma * np.sum(probs * U_old)
                else:
                    U_pi_k[s] = np.sum(probs * (rewards + gamma * U_old))

        # --- Policy Improvement ---
        policy_stable = True
        for s in mdp.S:
            old_action = pi_k[s]
            action_values = []
            for a in range(len(mdp.A)):
                probs = mdp.T[s, a]
                rewards = mdp.R[s, a]
                if np.ndim(rewards) == 0:
                    q = rewards + gamma * np.sum(probs * U_pi_k)
                else:
                    q = np.sum(probs * (rewards + gamma * U_pi_k))
                action_values.append(q)
            best_a = np.argmax(action_values)
            pi_kp1[s] = best_a
            if old_action != best_a:
                policy_stable = False

        # update policy for next iteration
        pi_k = copy.copy(pi_kp1)

        # optional: stop early if policy is stable
        if policy_stable:
            break

    return U_pi_k, pi_kp1
U_pi_k, pi_kp1 = policyIteration(mdp, numIterations=2)
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt


# Policy Iteration (Model-Based)
def policy_iteration(mdp, max_iter=1000, tol=1e-6):
    nS, nA = len(mdp.S), len(mdp.A)
    policy = np.zeros(nS, dtype=int)
    V = np.zeros(nS)
    iteration_times = []

    start = time.time()
    for it in range(max_iter):
        # Policy Evaluation
        while True:
            delta = 0
            for s in range(nS):
                v = V[s]
                a = policy[s]
                V[s] = np.sum(mdp.T[s, a, :] * (mdp.R[s, a] + mdp.discount * V))
                delta = max(delta, abs(v - V[s]))
            if delta < tol:
                break

        # Policy Improvement
        policy_stable = True
        for s in range(nS):
            old_action = policy[s]
            action_values = [np.sum(mdp.T[s, a, :] * (mdp.R[s, a] + mdp.discount * V)) for a in range(nA)]
            policy[s] = np.argmax(action_values)
            if old_action != policy[s]:
                policy_stable = False

        iteration_times.append(time.time() - start)
        if policy_stable:
            break

    total_time = time.time() - start
    return policy, V, iteration_times, total_time

# Q-Learning (Model-Free)
def q_learning(mdp, episodes=5000, alpha=0.1, gamma=0.9, epsilon=0.1):
    nS, nA = len(mdp.S), len(mdp.A)
    Q = np.zeros((nS, nA))
    rewards = []

    start = time.time()
    for ep in range(episodes):
        s = np.random.choice(mdp.S)
        total_reward = 0
        steps = 0
        while steps < 100:
            # epsilon-greedy action
            if np.random.rand() < epsilon:
                a = np.random.randint(nA)
            else:
                a = np.argmax(Q[s])

            # sample next state based on transition probabilities
            next_states = mdp.nextStates[(s,a)]
            if len(next_states) == 0:
                s_next = s  # stay in same state if no transitions
            else:
                probs = mdp.T[s, a, next_states]
                probs = probs / probs.sum()  # normalize just in case
                s_next = np.random.choice(next_states, p=probs)

            r = mdp.R[s,a]

            # Q update
            Q[s,a] = Q[s,a] + alpha * (r + gamma * np.max(Q[s_next]) - Q[s,a])

            total_reward += r
            s = s_next
            steps += 1

        rewards.append(total_reward)

    total_time = time.time() - start
    return Q, rewards, total_time

# ----------------------
# Run & Compare
# ----------------------
policy, V, policy_times, policy_time = policy_iteration(mdp)
Q, q_rewards, q_time = q_learning(mdp)

print(f"Policy Iteration: {policy_time:.4f} sec")
print(f"Q-Learning: {q_time:.4f} sec")

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(policy_times)
plt.title("Policy Iteration Convergence (time per iteration)")
plt.xlabel("Iteration")
plt.ylabel("Time (s)")

plt.subplot(1,2,2)
plt.plot(q_rewards)
plt.title("Q-Learning Rewards per Episode")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.show()
```

*   They learn directly from experience (trial and error with states, actions, rewards).

Model-Based Algorithms

*   Assume the environment dynamics are known (transition probabilities T and rewards R).
